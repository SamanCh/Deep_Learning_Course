# Loss Functions

## MSE

```python
import numpy as np #
import matplotlib.pyplot as plt

# Define the MSE loss function
def mean_squared_error(y_true, y_pred):
    """
    Calculate the Mean Squared Error between true and predicted values.
    
    Parameters:
    y_true (numpy array): Actual values
    y_pred (numpy array): Predicted values
    
    Returns:
    float: The MSE value
    """
    mse = np.mean((y_true - y_pred) ** 2)
    return mse

# Example usage
# Actual values (ground truth)
y_true = np.array([3.0, -0.5, 2.0, 7.0])

# Predicted values by the model
y_pred = np.array([2.5, 0.0, 2.0, 8.0])

# Calculate the MSE
mse_value = mean_squared_error(y_true, y_pred)
print("Mean Squared Error:", mse_value)

# Plotting the squared errors
squared_errors = (y_true - y_pred) ** 2
plt.figure(figsize=(10, 6))

plt.bar(range(len(y_true)), squared_errors, color='skyblue', alpha=0.6)
plt.axhline(y=mse_value, color='r', linestyle='--', label=f'MSE = {mse_value:.3f}')
plt.xticks(range(len(y_true)), labels=[f'Point {i+1}' for i in range(len(y_true))])
plt.xlabel('Data Points')
plt.ylabel('Squared Error')
plt.title('Squared Errors for Each Data Point')
plt.legend()

plt.show()
```

## Cross Entropy

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Cross-Entropy Loss function for binary classification
def cross_entropy_loss(y_true, y_pred):
    """
    Calculate the Cross-Entropy Loss for binary classification.
    
    Parameters:
    y_true (numpy array): Actual binary labels (0 or 1)
    y_pred (numpy array): Predicted probabilities
    
    Returns:
    float: The Cross-Entropy Loss value
    """
    epsilon = 1e-15  # to avoid log(0)
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)  # clip predictions to avoid log(0)
    cross_entropy = -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
    return cross_entropy

# Example usage
# Actual labels (ground truth)
y_true = np.array([1, 0, 1, 0])

# Predicted probabilities by the model
y_pred = np.array([0.9, 0.1, 0.8, 0.4])

# Calculate the Cross-Entropy Loss
cross_entropy_value = cross_entropy_loss(y_true, y_pred)
print("Cross-Entropy Loss:", cross_entropy_value)

# Plotting the Cross-Entropy Loss for a range of predicted probabilities
predicted_probs = np.linspace(0.01, 0.99, 100)
loss_1 = [-np.log(p) for p in predicted_probs]  # Loss when y_true = 1
loss_0 = [-np.log(1 - p) for p in predicted_probs]  # Loss when y_true = 0

plt.figure(figsize=(10, 6))
plt.plot(predicted_probs, loss_1, label="True Label = 1", color='blue')
plt.plot(predicted_probs, loss_0, label="True Label = 0", color='green')
plt.xlabel("Predicted Probability")
plt.ylabel("Cross-Entropy Loss")
plt.title("Cross-Entropy Loss for True Labels 1 and 0")
plt.legend()
plt.show()
```

## Softmax

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Softmax function
def softmax(z):
    """
    Compute the Softmax of a vector z.
    
    Parameters:
    z (numpy array): Input vector (logits)
    
    Returns:
    numpy array: Softmax probabilities
    """
    exp_z = np.exp(z - np.max(z))  # Subtract max(z) for numerical stability
    return exp_z / np.sum(exp_z)

# Example usage
# Input logits (raw scores)
z = np.array([1.0, 2.0, 3.0])

# Calculate the Softmax probabilities
softmax_probs = softmax(z)
print("Softmax Probabilities:", softmax_probs)

# Plotting the Softmax function for varying logits
logits = np.linspace(-2.0, 2.0, 100)
softmax_vals = np.array([softmax([l, 0, 0]) for l in logits])

plt.figure(figsize=(10, 6))
plt.plot(logits, softmax_vals[:, 0], label='Softmax of logit 1')
plt.plot(logits, softmax_vals[:, 1], label='Softmax of logit 2')
plt.plot(logits, softmax_vals[:, 2], label='Softmax of logit 3')
plt.xlabel("Logit Value")
plt.ylabel("Softmax Probability")
plt.title("Softmax Function Applied to Logits")
plt.legend()
plt.show()
```
