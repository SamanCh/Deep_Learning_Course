# Activation functions

The activation function in an artificial neural network (ANN) plays a critical role in determining the output of a neuron. It introduces non-linearity into the model, which is essential for the network to learn and model complex patterns.

## AND Gate to an example of activation functions:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the inputs and expected outputs (AND gate)
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])  # Expected outputs

# Initialize weights and bias
weights = np.zeros(X.shape[1])
bias = 0
learning_rate = 0.1
epochs = 10

# Perceptron training function
def train(X, y, weights, bias, learning_rate, epochs):
    for _ in range(epochs):
        for i in range(len(y)):
            # Weighted sum
            linear_output = np.dot(X[i], weights) + bias
            # Apply step function (activation)
            y_pred = 1 if linear_output > 0 else 0
            # Update weights and bias
            weights += learning_rate * (y[i] - y_pred) * X[i]
            bias += learning_rate * (y[i] - y_pred)
    return weights, bias

# Train the perceptron
weights, bias = train(X, y, weights, bias, learning_rate, epochs)

# Create a meshgrid for plotting the decision boundary
xx, yy = np.meshgrid(np.linspace(-0.5, 1.5, 400), np.linspace(-0.5, 1.5, 400))
Z = np.dot(np.c_[xx.ravel(), yy.ravel()], weights) + bias
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.figure(figsize=(6, 6))
plt.contourf(xx, yy, Z > 0, alpha=0.4, levels=[0, 1], colors=['red', 'green'])
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', s=100, cmap=plt.cm.Paired)

# Set labels and title
plt.xlabel('Input A')
plt.ylabel('Input B')
plt.title('Perceptron as a Linear Classifier for AND Gate')

# Set axis limits and grid
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the derivative of the sigmoid function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Generate values for plotting
x_values = np.linspace(-10, 10, 400)
y_values = sigmoid(x_values)
y_derivative = sigmoid_derivative(x_values)

# Plot the sigmoid function and its derivative
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Sigmoid Function')
plt.plot(x_values, y_derivative, label='Derivative of Sigmoid', linestyle='--')

# Add labels and title
plt.xlabel('x')
plt.ylabel('Sigmoid(x) / Sigmoid\'(x)')
plt.title('Sigmoid Function and its Derivative')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

## Hyperbolic Tangent Function and Its Derivative

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the tanh function
def tanh(x):
    return np.tanh(x)

# Define the derivative of tanh function
def tanh_derivative(x):
    return 1.0 - np.tanh(x)**2

# Generate data
x = np.linspace(-5, 5, 400)
y = tanh(x)

# Plot the tanh function
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='tanh(x)', color='b')

# Plot the derivative of the tanh function
y_derivative = tanh_derivative(x)
plt.plot(x, y_derivative, '--', label="tanh'(x)", color='r')

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Hyperbolic Tangent Function and Its Derivative')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

## ReLu Function and Its Derivative

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ReLU function
def relu(x):
    return np.maximum(0, x)

# Define the derivative of the ReLU function
def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# Generate input values
x = np.linspace(-10, 10, 400)

# Compute ReLU activation and its derivative
y = relu(x)
dy = relu_derivative(x)

# Create the plot for ReLU
plt.figure(figsize=(12, 6))

# Plot ReLU function
plt.subplot(1, 2, 1)
plt.plot(x, y, label='ReLU', color='blue')
plt.title('ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()

# Plot ReLU derivative function
plt.subplot(1, 2, 2)
plt.plot(x, dy, label="ReLU's Derivative", color='red')
plt.title("ReLU's Derivative")
plt.xlabel('Input')
plt.ylabel('Derivative')
plt.grid(True)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

## Leaky ReLU Function and Its Derivative

```python
import numpy as np
import matplotlib.pyplot as plt

# Leaky ReLU function
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Derivative of Leaky ReLU function
def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

# Generate input values
x = np.linspace(-10, 10, 400)

# Apply Leaky ReLU function and its derivative
y = leaky_relu(x, alpha=0.01)
dy = leaky_relu_derivative(x, alpha=0.01)

# Plotting the Leaky ReLU function
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, y, label="Leaky ReLU (=0.01)", color='b')
plt.title("Leaky ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

# Plotting the derivative of Leaky ReLU function
plt.subplot(1, 2, 2)
plt.plot(x, dy, label="Derivative of Leaky ReLU (=0.01)", color='r')
plt.title("Derivative of Leaky ReLU")
plt.xlabel("Input")
plt.ylabel("Derivative")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

plt.show()
```
